```python
class LinearRegression:
  def __init__(self):
    self._a = None
    self._b = None
    self._x_avg = None
    self._y_avg = None

  def train_test_partition(self, x, y):
    x_train, x_test = [], []
    y_train, y_test = [], []
    for i in range(int(len(x) * 0.75)):
      x_train.append(x[i])
      y_train.append(y[i])
    for i in range(int(len(x) * 0.75), len(x)):
      x_test.append(x[i])
      y_test.append(y[i])

  def fit(self, x, y):
    self._x_avg = sum(x)/len(x)
    self._y_avg = sum(y)/len(y)
    num, den = 0, 0
    for i, x_point in enumerate(x):
      num += (x_point - self._x_avg) * (y[i] - self._y_avg)
      den += (x_point - self._x_avg) ** 2
    self._a = num / den
    self._b = self._y_avg - self._a * self._x_avg

  def predict(self, x_test):
    if self._a is None or self._b is None: return None
    if isinstance(x_test, list):
      return [self._a * val + self._b for val in x_test]
    else:
      return self._a * x_test + self._b

  def get_coords(self):
    return f'gradient = {self._a}, y-inercept = {self._b}'
```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/canada_per_capita_income.csv')
df.head()
```
